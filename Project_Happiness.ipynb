```python
#Look for unique region in column 'Region' to prepare a variable to be used as a criteria for record selection and create new array.
Region_list = df['Region'].unique()
#Array(['Western Europe', 'Africa', 'Eastern Europe', 'Asia-Pacific','Latin America', 'North America', 'Europe'], dtype=object)
#Obtain an array with all the region name.
```

```python
#Loop for time the number of region listed in the Region_list array. Select the all data points for each region listed in the Region_list array in df_region new dataframe and record the corresponding .csv file.
for i in range(len(Region_list)):
    df_region = df[df['Region']== str(Region_list[i])]
    df_region.to_csv(str(Region_list[i])+'.csv', index=False)
```

```python
#Creare an empty dataframe to store the result of Region happiness ranking data, score > 6 and numbwe of country with score above 6.
df_region_happiness = pd.DataFrame(columns=['Region','Mean Happiness','# Country above 6','Delta Range'])
#loop to read the region csv files created previously as asked in the exercise.
for i in range(len(Region_list)):
    Region_each = pd.read_csv(Region_list[i]+'.csv')
    #Fill the columns of empty dataframe with Region and Mean happiness
    df_region_happiness = df_region_happiness.append({'Region': Region_list[i], 'Mean Happiness': str(Region_each['Happiness Score'].mean()), '# Country above 6': int(len(Region_each[Region_each['Happiness Score'] > 6])), 'Delta Range': Region_each['Happiness Score'].max()-Region_each['Happiness Score'].min()}, ignore_index=True)     
#Change type of column Mean Happiness to numerical to allow ranking
df_region_happiness['Mean Happiness'] = df_region_happiness['Mean Happiness'].astype('float64')
df_region_happiness.sort_values(by = 'Mean Happiness', ascending = False)
```

```python
#Select and show the row where the delta range is the largest.
df_region_happiness[df_region_happiness['Delta Range'] == df_region_happiness['Delta Range'].max() ]
```

```python
#Create specific dataframe for the 10 happiest country for the exercise.
df_ten_happiest = df.sort_values(by = ['Happiness Rank']).head(10)
#Make the dataframe in descending order because I want this to be displayed in descending on the horizontal bar graph.
df_ten_happiest = df_ten_happiest.sort_values(by =['Happiness Rank'], ascending=False).copy()
```

```python
plt.barh(df_ten_happiest['Country'], df_ten_happiest['Happiness Score'])
plt.xlim(0,8)
#Put title and X,Y label on the figure.
plt.title('Ten Happiest Countries')
plt.ylabel('Country')
plt.xlabel('Happines Score')
```

```python
#Task E with linear regression displayed (use of kind='reg' argument)
sns.pairplot(df, x_vars=['Happiness Score'], y_vars=df.columns[3:11], kind='reg')
#The best visible correlation is Job Score with Economy but it will be necessary to calculate.
```

```python
#Create an empty dataframe to store the result of the Happiness Score correlation.
df_correlation = pd.DataFrame(columns=['Category','Correlation Coeficient'])
#Loop for each category and calculate successively the correlation coefficient with Happiness Score and store in the df_correlation dataframe.
for col_name in df.columns[3:11]:
    df_correlation= df_correlation.append({'Category': str(col_name), 'Correlation Coeficient': df['Happiness Score'].corr(df[col_name])},ignore_index=True)
df_correlation.sort_values(by =['Correlation Coeficient'], ascending=False)
#The highest correlation with Happiness Score is Job Satifaction with 0.812873
```
